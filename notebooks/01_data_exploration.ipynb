# MediSense — Dataset Exploration
Load and explore all Kaggle datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

DATA_DIR = '../data/symptom_disease/'
print('Files in data dir:', os.listdir(DATA_DIR) if os.path.exists(DATA_DIR) else 'No files yet')
```

```python
# Load disease-symptom dataset
# Download first: kaggle datasets download -d itachi9604/disease-symptom-description-dataset
try:
    df = pd.read_csv(DATA_DIR + 'dataset.csv')
    print(f'Rows: {len(df)}, Columns: {len(df.columns)}')
    print(df.head())
    print('\nDiseases:', df['Disease'].nunique())
    print(df['Disease'].value_counts().head(10))
except FileNotFoundError:
    print('Download dataset first! See README for kaggle CLI commands.')
```

```python
# Visualize symptom distribution
symptom_cols = [c for c in df.columns if 'Symptom' in c]
all_symptoms = pd.Series([val for col in symptom_cols for val in df[col].dropna().values])
print('Top 20 most common symptoms:')
print(all_symptoms.value_counts().head(20))
```
